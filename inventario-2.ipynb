```python
import csv
import os
from datetime import datetime

class Producto:
    def __init__(self, id, nombre, categoria, precio, stock):
        self.id = str(id)
        self.nombre = nombre.strip()
        self.categoria = categoria.strip()
        self.precio = self._validar_precio(precio)
        self.stock = self._validar_stock(stock)

    def _validar_precio(self, precio):
        precio = float(precio)
        if precio <= 0:
            raise ValueError("El precio debe ser mayor a 0")
        return precio

    def _validar_stock(self, stock):
        stock = int(stock)
        if stock < 0:
            raise ValueError("El stock no puede ser negativo")
        return stock

    def esta_agotado(self):
        return self.stock == 0

    def stock_bajo(self, minimo=5):
        return self.stock <= minimo

    def __str__(self):
        estado = "🔴 AGOTADO" if self.esta_agotado() else "🟡 BAJO" if self.stock_bajo() else "🟢 OK"
        return f"{self.id:<6} | {self.nombre:<20} | {self.categoria:<15} | ${self.precio:>8.2f} | Stock: {self.stock:>4} {estado}"

print("✅ Clase Producto cargada")
```

```python
class Inventario:
    def __init__(self, archivo='inventario.csv'):
        self.archivo = archivo
        self.productos = {}
        self._cargar()

    def _cargar(self):
        if not os.path.exists(self.archivo):
            self._crear_csv()
            return
        with open(self.archivo, 'r') as f:
            reader = csv.DictReader(f)
            for row in reader:
                try:
                    p = Producto(row['id'], row['nombre'],
                                row['categoria'], row['precio'], row['stock'])
                    self.productos[p.id] = p
                except ValueError as e:
                    print(f"⚠️ Fila ignorada: {e}")
        print(f"✅ {len(self.productos)} productos cargados")

    def _crear_csv(self):
        with open(self.archivo, 'w', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(['id', 'nombre', 'categoria', 'precio', 'stock'])
        print(f"✅ Archivo {self.archivo} creado")

    def _guardar_todo(self):
        with open(self.archivo, 'w', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(['id', 'nombre', 'categoria', 'precio', 'stock'])
            for p in self.productos.values():
                writer.writerow([p.id, p.nombre, p.categoria, p.precio, p.stock])

    def agregar(self, nombre, categoria, precio, stock):
        nuevo_id = str(len(self.productos) + 1).zfill(3)
        p = Producto(nuevo_id, nombre, categoria, precio, stock)
        self.productos[nuevo_id] = p
        self._guardar_todo()
        print(f"✅ Producto agregado:\n{p}")
        return p

    def actualizar_stock(self, id, cantidad):
        if id not in self.productos:
            raise KeyError(f"Producto {id} no encontrado")
        self.productos[id].stock += cantidad
        self._guardar_todo()
        print(f"✅ Stock actualizado: {self.productos[id]}")

    def buscar(self, termino):
        termino = termino.lower()
        resultados = [p for p in self.productos.values()
                     if termino in p.nombre.lower() or termino in p.categoria.lower()]
        return resultados

    def stock_bajo(self):
        return [p for p in self.productos.values() if p.stock_bajo()]

    def reporte(self):
        if not self.productos:
            print("No hay productos registrados.")
            return

        separador = "=" * 65
        print(f"\n{separador}")
        print(f"{'📦 REPORTE DE INVENTARIO':^65}")
        print(separador)
        print(f"{'ID':<6} | {'NOMBRE':<20} | {'CATEGORÍA':<15} | {'PRECIO':>9} | {'STOCK':>9}")
        print("-" * 65)
        for p in self.productos.values():
            print(p)

        valor_total = sum(p.precio * p.stock for p in self.productos.values())
        print(f"\n  Total productos: {len(self.productos)}")
        print(f"  Valor total inventario: ${valor_total:,.2f}")

        bajos = self.stock_bajo()
        if bajos:
            print(f"\n⚠️  PRODUCTOS CON STOCK BAJO ({len(bajos)}):")
            for p in bajos:
                print(f"   → {p.nombre} — Stock: {p.stock}")
        print(f"\n{separador}\n")

print("✅ Clase Inventario cargada")
```

```python
import csv       # Para leer y escribir archivos CSV
import os        # Para verificar si un archivo existe
from datetime import datetime  # Para manejar fechas si las necesitamos

class Producto:
    def __init__(self, id, nombre, categoria, precio, stock):
        # self representa el objeto actual — cada producto tiene sus propios valores
        self.id = str(id)              # Convertimos a string para estandarizar
        self.nombre = nombre.strip()   # .strip() elimina espacios al inicio y final
        self.categoria = categoria.strip()
        self.precio = self._validar_precio(precio)   # Validamos antes de guardar
        self.stock = self._validar_stock(stock)      # Validamos antes de guardar

    def _validar_precio(self, precio):
        # El guión bajo _ indica que es método privado (solo uso interno)
        precio = float(precio)   # Convertimos a número decimal
        if precio <= 0:
            raise ValueError("El precio debe ser mayor a 0")  # Lanzamos error si es inválido
        return precio  # Si todo está bien, devolvemos el precio validado

    def _validar_stock(self, stock):
        stock = int(stock)   # Stock siempre es número entero
        if stock < 0:
            raise ValueError("El stock no puede ser negativo")
        return stock

    def esta_agotado(self):
        # Devuelve True si stock es 0, False si hay existencias
        return self.stock == 0

    def stock_bajo(self, minimo=5):
        # minimo=5 es el valor por defecto si no pasas otro número
        return self.stock <= minimo

    def __str__(self):
        # __str__ define cómo se ve el objeto cuando haces print(producto)
        # Determinamos el emoji de estado según el stock
        estado = "🔴 AGOTADO" if self.esta_agotado() else "🟡 BAJO" if self.stock_bajo() else "🟢 OK"
        # :<6 significa alineado a la izquierda con 6 caracteres de espacio
        # :>8.2f significa alineado a la derecha, 8 caracteres, 2 decimales
        return f"{self.id:<6} | {self.nombre:<20} | {self.categoria:<15} | ${self.precio:>8.2f} | Stock: {self.stock:>4} {estado}"

print("✅ Clase Producto cargada")
```

```python
class Inventario:
    def __init__(self, archivo='inventario.csv'):
        self.archivo = archivo       # Guardamos el nombre del archivo
        self.productos = {}          # Diccionario vacío: {id: objeto Producto}
        self._cargar()               # Cargamos productos al iniciar

    def _cargar(self):
        # Si el archivo no existe, lo creamos desde cero
        if not os.path.exists(self.archivo):
            self._crear_csv()
            return  # Salimos de la función aquí

        # Si ya existe, leemos cada fila y creamos objetos Producto
        with open(self.archivo, 'r') as f:
            reader = csv.DictReader(f)  # DictReader convierte cada fila en diccionario
            for row in reader:
                try:
                    # Creamos un Producto con los datos de cada fila
                    p = Producto(row['id'], row['nombre'],
                                row['categoria'], row['precio'], row['stock'])
                    # Lo guardamos en el diccionario usando su ID como llave
                    self.productos[p.id] = p
                except ValueError as e:
                    # Si una fila tiene datos inválidos, la ignoramos y seguimos
                    print(f"⚠️ Fila ignorada: {e}")
        print(f"✅ {len(self.productos)} productos cargados")

    def _crear_csv(self):
        # Creamos el archivo con solo el encabezado
        with open(self.archivo, 'w', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(['id', 'nombre', 'categoria', 'precio', 'stock'])
        print(f"✅ Archivo {self.archivo} creado")

    def _guardar_todo(self):
        # Reescribimos el CSV completo cada vez que hay un cambio
        with open(self.archivo, 'w', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(['id', 'nombre', 'categoria', 'precio', 'stock'])
            for p in self.productos.values():
                # .values() nos da todos los objetos Producto del diccionario
                writer.writerow([p.id, p.nombre, p.categoria, p.precio, p.stock])

    def agregar(self, nombre, categoria, precio, stock):
        # Generamos ID automático: 001, 002, 003...
        # zfill(3) rellena con ceros a la izquierda hasta 3 dígitos
        nuevo_id = str(len(self.productos) + 1).zfill(3)
        p = Producto(nuevo_id, nombre, categoria, precio, stock)
        self.productos[nuevo_id] = p   # Lo agregamos al diccionario en memoria
        self._guardar_todo()            # Lo guardamos en el CSV
        print(f"✅ Producto agregado:\n{p}")
        return p

    def actualizar_stock(self, id, cantidad):
        # Verificamos que el ID existe antes de modificar
        if id not in self.productos:
            raise KeyError(f"Producto {id} no encontrado")
        # += suma si es positivo (entrada), resta si es negativo (salida)
        self.productos[id].stock += cantidad
        self._guardar_todo()
        print(f"✅ Stock actualizado: {self.productos[id]}")

    def buscar(self, termino):
        termino = termino.lower()  # Convertimos a minúsculas para buscar sin importar mayúsculas
        # List comprehension: filtra productos donde el término aparece en nombre o categoría
        resultados = [p for p in self.productos.values()
                     if termino in p.nombre.lower() or termino in p.categoria.lower()]
        return resultados

    def stock_bajo(self):
        # Devuelve lista de productos con stock menor o igual a 5
        return [p for p in self.productos.values() if p.stock_bajo()]

    def reporte(self):
        if not self.productos:
            print("No hay productos registrados.")
            return

        separador = "=" * 65
        print(f"\n{separador}")
        print(f"{'📦 REPORTE DE INVENTARIO':^65}")  # :^65 centra el texto en 65 caracteres
        print(separador)
        print(f"{'ID':<6} | {'NOMBRE':<20} | {'CATEGORÍA':<15} | {'PRECIO':>9} | {'STOCK':>9}")
        print("-" * 65)

        # Imprimimos cada producto usando su __str__
        for p in self.productos.values():
            print(p)

        # Calculamos valor total: precio × stock de cada producto
        valor_total = sum(p.precio * p.stock for p in self.productos.values())
        print(f"\n  Total productos: {len(self.productos)}")
        print(f"  Valor total inventario: ${valor_total:,.2f}")  # , agrega separador de miles

        # Alertas de stock bajo
        bajos = self.stock_bajo()
        if bajos:
            print(f"\n⚠️  PRODUCTOS CON STOCK BAJO ({len(bajos)}):")
            for p in bajos:
                print(f"   → {p.nombre} — Stock: {p.stock}")
        print(f"\n{separador}\n")

print("✅ Clase Inventario cargada")
```
